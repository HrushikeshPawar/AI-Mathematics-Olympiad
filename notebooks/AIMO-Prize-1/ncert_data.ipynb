## Token Counting

````python
import tiktoken
def get_tokens_openai(input_text, model="gpt-3.5-turbo-0613"):
    """Return the number of tokens used by a list of messages."""
    encoding = tiktoken.encoding_for_model(model)
    tokens_per_message = 3
    return tokens_per_message + len(encoding.encode(input_text)) + 3
````

````python
problems = class_6_df['problem'].tolist() + class_7_df['problem'].tolist() + class_8_df['problem'].tolist() + class_9_df['problem'].tolist() + class_10_df['problem'].tolist() + class_11_df['problem'].tolist() + class_12_df['problem'].tolist()

solutions = class_6_df['solution'].tolist() + class_7_df['solution'].tolist() + class_8_df['solution'].tolist() + class_9_df['solution'].tolist() + class_10_df['solution'].tolist() + class_11_df['solution'].tolist() + class_12_df['solution'].tolist()
````

````python
len(problems), len(solutions)
````

````python
prob_token_counts = []
for problem in tqdm(problems):
    # try:
    #     prob_token_counts.append(num_tokens_from_messages(problem))
    # except AttributeError:
    #     print(problem[0])
    #     break
    prob_token_counts.append(get_tokens_openai(problem[0]))
````

````python
sys_prompt = """You are a math professor who is very rigorous in writing down solutions. You also care about your students and write solution walkthrough in step-by-step manner and using latex everywhere necessary.

You will be given a math problem and its solution, you job is to re-write the problem in good latex and solution in very step-by-step manner for your students.
If you can find an efficient solution than given, please give that solution.

The output has to be in the YAML format:
```
to_drop: BOOLEAN
problem: "PROBLEM RE-WRITTEN IN PROPER LATEX"
solution:
  - "SOLUTION RE-WRITTEN IN PROPER LATEX IN MULTIPLE STEPS AS A LIST"
```
You should drop a problem altogether if its incomplete or problem has includes a diagram or is about constructing one or the problem is MCQ.
Keep the number of steps as minimum as possible.

Study the given solution, if you can solve the problem with a more efficient solution give that.
Combine latex commands in single step, not multiple.

DO NOT USE `\\begin{align*}` `\\begin{align}` `\\begin{aligned}` or any kind of other alignment in latex code.
Take care that you use right YAML format, do not nest steps!
"""
````

````python
input_prompt = """<math_problem>
{math_problem}
</math_problem>

<solution>
{solution}
</solution>"""
````

````python
exp_id = 0
jsonl_fnames = [f'cleaned_class_{class_num}.jsonl' for class_num in range(6, 13)]
ids = []
problems = []
solutions = []
prompt = []
concepts = []
chapter_names = []

for jsonl_fname in jsonl_fnames:
    var_df = pd.read_json(os.path.join(BASE_FOLDER, jsonl_fname), lines=True)
    for idx in tqdm(range(len(var_df))):
        prob = "\n".join(var_df.problem[idx])
        sols = ""
        for sol in var_df.solution[idx]:
            sols += "\n".join(sol) + "\n\n"
        sols = sols[:-2]
        input_text = input_prompt.format(math_problem=prob, solution=sols)
        
        ids.append(f"Exp{exp_id:0>4}")
        problems.append(prob)
        solutions.append(sols)
        prompt.append(input_text)
        concepts.append(var_df.concept[idx])
        chapter_names.append(var_df.chapter_name[idx])

        var_df.loc[idx, 'exp_id'] = f"Exp{exp_id:0>4}"
        exp_id += 1
    
    # var_df.to_json(os.path.join(BASE_FOLDER, jsonl_fname), index=False, lines=True, orient='records')
````

````python
preprocess_df = pd.DataFrame({
    "id": ids,
    "problem": problems,
    "solution": solutions,
    "prompt": prompt,
    "concept": concepts,
    "chapter_name": chapter_names
})
preprocess_df
````

````python
preprocess_df.to_json(os.path.join(BASE_FOLDER, 'preprocess_input.jsonl'), index=False, lines=True, orient='records')
````
