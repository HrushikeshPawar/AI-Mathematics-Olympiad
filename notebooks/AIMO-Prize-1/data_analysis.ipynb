```python
idx = 46
folder_path = os.path.join(DATA_DIR, 'amps', 'khan', folders[idx])
files = os.listdir(folder_path)
drop_files = []

with open(os.path.join(folder_path, files[0])) as f:
    data = json.load(f)
display(Markdown(f"### Folder: {folders[idx]}\n\n### Problem:\n\n{data['problem']}\n\n### Hints:\n\n{' '.join(data['hints'])}".replace('align*', 'aligned')))
```

```python
checking_term = r'using substitution'.lower()
flag = False
for file in files:
    with open(os.path.join(folder_path, file)) as f:
        data = json.load(f)
    if checking_term not in data['problem'].lower():
        print(f'Not found: file={file} problem={data["problem"]}\n')
        # print(f'Not found: file={file}')
        drop_files.append(file)
        flag = True

if not flag:
    print('All files have the term')
```
